```python
cols = [
    'area_sqft',
    'age_years',
    'distance_city_km',
    'lot_size_sqft',
    'renovation_years_ago',
    'house_price_inr'
]

df_clean = df.copy()

for col in cols:

    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_clean[col] = np.where(
        df_clean[col] < lower,
        lower,
        df_clean[col]
    )

    df_clean[col] = np.where(
        df_clean[col] > upper,
        upper,
        df_clean[col]
    )
```

This code detects outliers in six numerical features using the IQR method. For each feature, the first quartile (Q1), third quartile (Q3), and Interquartile Range (IQR) are calculated. Lower and upper limits are then determined using the formula Q1 − 1.5×IQR and Q3 + 1.5×IQR. Instead of deleting outliers, extreme values are replaced with the nearest boundary values. This process reduces the impact of outliers while preserving all records in the dataset, making the data more suitable for machine learning model training.

```python
numerical_cols = [
    'area_sqft',
    'age_years',
    'distance_city_km',
    'lot_size_sqft',
    'renovation_years_ago',
    'house_price_inr'
]

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col}: {len(outliers)} outliers")
```

This code identifies the number of outliers present in each numerical feature using the Interquartile Range (IQR) method. For every numerical column, the first quartile (Q1), third quartile (Q3), and IQR are calculated. Based on these values, lower and upper boundaries are determined using the formulas Q1 − 1.5 × IQR and Q3 + 1.5 × IQR. Any data point falling outside these boundaries is considered an outlier. The code then counts and displays the total number of outliers for each feature. This analysis helps understand the extent of extreme values in the dataset before applying any outlier treatment or machine learning techniques.

```python
X_simple = df_clean[['area_sqft']]

y_simple = df_clean['house_price_inr']
```

This step prepares the dataset for Simple Linear Regression. Only one feature, area_sqft, is selected as the independent variable because house area is often one of the most influential factors affecting property prices. The target variable is house_price_inr, which represents the house price to be predicted. The relationship between house area and house price will be analyzed using a simple linear regression model.

### Simple Linear Regression

```python
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_simple,
    y_simple,
    test_size=0.20,
    random_state=42
)
```

This step divides the simple regression dataset into training and testing subsets. The training set contains 80% of the observations and is used to train the Simple Linear Regression model. The testing set contains the remaining 20% of observations and is used to evaluate the model's predictive performance on unseen data. Using a fixed random state ensures consistent and reproducible results.

```python
from sklearn.linear_model import LinearRegression

slr = LinearRegression()

slr.fit(X_train_s, y_train_s)
```

This step creates and trains a Simple Linear Regression model using house area as the independent variable and house price as the dependent variable. The model learns the relationship between these variables from the training data and generates a mathematical equation that can be used to predict house prices based on the area of a house.

```python
pred_simple = slr.predict(X_test_s)
```

This step generates house price predictions using the trained Simple Linear Regression model. The model applies the relationship learned during training to the unseen testing data and estimates the house prices based on house area. These predicted values will later be compared with the actual house prices to evaluate the model's accuracy and performance.

House Price=m×Area+c
m = Slope (coefficient)
c = Intercept

```python
print("Slope:", slr.coef_[0])

print("Intercept:", slr.intercept_)
```


This code retrieves the parameters of the trained Simple Linear Regression model. The slope represents the change in house price for each additional square foot of area, while the intercept represents the baseline predicted price when the area is zero. Together, these values form the regression equation used for house price prediction.

House Price=(Slope×Area)+Intercept

Price=(4850.75×1000)+250000.50
Price=5,100,750.50

```python
plt.figure(figsize=(8,5))

plt.scatter(
    X_test_s,
    y_test_s,
    alpha=0.5
)

plt.plot(
    X_test_s,
    pred_simple,
    color='red'
)

plt.xlabel("Area (sqft)")
plt.ylabel("House Price")
plt.title("Simple Linear Regression")

plt.show()
```

## This code is used to visualize the Best Fit Line of the Simple Linear Regression model.

This code visualizes the results of the Simple Linear Regression model. The scatter plot represents the actual house prices from the testing dataset, while the red line represents the predicted prices generated by the regression model. The graph helps evaluate how well the regression line fits the actual data and illustrates the relationship between house area and house price.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse_simple = mean_squared_error(
    y_test_s,
    pred_simple
)

mae_simple = mean_absolute_error(
    y_test_s,
    pred_simple
)

rmse_simple = np.sqrt(mse_simple)

r2_simple = r2_score(
    y_test_s,
    pred_simple
)

print("MSE :", mse_simple)
print("MAE :", mae_simple)
print("RMSE :", rmse_simple)
print("R2 Score :", r2_simple)
```

This code evaluates the performance of the Simple Linear Regression model using four commonly used regression metrics: Mean Squared Error (MSE), Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and R² Score. These metrics measure the difference between actual and predicted house prices and indicate how accurately the model performs on unseen test data. Lower error values and a higher R² Score indicate better model performance.

MSE stands for Mean Squared Error, MAE stands for Mean Absolute Error, RMSE stands for Root Mean Squared Error, and R² Score stands for Coefficient of Determination. These metrics are used to evaluate the performance and accuracy of a Linear Regression model.

```python
n = len(y_test_s)

p = 1

adj_r2_simple = 1 - (
    ((1-r2_simple)*(n-1))
    /
    (n-p-1)
)

print("Adjusted R2 :", adj_r2_simple)
```

This code calculates the Adjusted R² Score of the Simple Linear Regression model. Adjusted R² measures how well the independent variable explains the variation in the dependent variable while accounting for the number of predictors used in the model. It provides a more reliable evaluation metric than R² when comparing regression models with different numbers of features.

R² Score (Coefficient of Determination) measures how well the model explains the variation in the target variable. A value closer to 1 indicates a better model.

Adjusted R² Score (Adjusted Coefficient of Determination) measures model performance while considering the number of features used. A higher value indicates a better and more reliable model.
